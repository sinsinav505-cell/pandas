```python
import pandas as pd
data = [1,2,3,4,'apple']

#convert the list into panda series
series = pd.Series(data, index=[101,102,103,104,105])
print(series)
```

```python

weeks={
    "Day1":"Monday",
    "Day2":"Tuesday",
    "Day3":"Wednesday",
    "Day4":"Thursday",
    "Day5":"Friday",
    "Day6":"Saturday",
    "Day7":"Sunday"
}

day_series = pd.Series(weeks)
print(day_series)
print("\n Series created from a dictionary")
print("\n First element using iloc:")
print(day_series.iloc[0])
```

```python

data = [1,2,3,4,5]

#convert the list into panda series
series = pd.Series(data, index=[101,102,103,104,105])
print(series[series>3])
```

```python
##  Series from a list with filtering
# -----------------------------------------------------
# Task:
# - Create a Series from the list [10, 20, 30, 40, 50]
# - Use loc[] to access the element with index 3
# - Use iloc[] to access the second element
# - Filter and print elements greater than 25

l = [10,20,30,40,50]
a = pd.Series(l)
print(a.loc[3])
print(a.iloc[1]) 
print(a[a>25])



```

```python
#  Series from a dictionary
# -----------------------------------------------------
# Task:
# - Create a Series from the dictionary: {"A": 5, "B": 15, "C": 25, "D": 35}
# - Access the value of key "B" using loc[]
# - Access the first element using iloc[]
# - Filter and print values greater than 20

dic = {"A":5,"B":15,"C":25,"D":35}
b = pd.Series(dic)
print(b.loc["B"])
print(b.iloc[0])
print(b[b>20])
```

```python
# DataFrame from a dictionary
# -----------------------------------------------------
# Task:
# - Create a DataFrame with columns: "Name", "Age", "City"
#   Data: ["Alice", "Bob", "Charlie"], [25, 30, 35], ["Delhi", "Mumbai", "Chennai"]
# - Print the entire DataFrame
# - Print the row at index 1 using loc[]
# - Print the first row using iloc[]

t = { "Name":["Alice","Bob","Charlie"],"Age":[25,30,35],"City":["Delhi","Mumbai","Chennai"]}
c = pd.DataFrame(t)
print(c)
print(c.loc[1])
print(c.iloc[0])


#  Conditional filtering in DataFrame
# -----------------------------------------------------
# Task:
# - Using the above DataFrame, filter rows where Age > 28
# - Filter rows where City is "Delhi"
# - Print the filtered DataFrames

print(c[c["Age"]>28])
print(c[c["City"]=="Delhi"])

```
