```python
import pandas as pd
```

```python
a=pd.read_csv('TV_Final.csv')
print(a.to_string())
```

```python
# Q1: Display all rows where the Brand is "SAMSUNG" AND the Selling Price is less than 50000.

a[(a['Brand'] == "SAMSUNG") & (a['Selling Price']<50000)]
```

```python
# Q2: Display only the Brand, Size, and Rating columns for HD LED
#  where the Operating System is "Android " AND Rating is greater than or equal to 4.0.

a[(a['Operating System'] == "Android") & (a['Rating'] >= 4.0)][['Brand','Size','Rating']]
```

```python
#  Drop rows with missing values
# Q3: Drop all rows where the Rating column has NaN values. 

a.dropna(subset=['Rating'])
```

```python
# Fill missing values
# Q4: Replace all NaN values in the Operating System column with the string "Unknown"

a.fillna({"Operating System":'Unknown'}, inplace=False)
```

```python
#  Rename columns
# Q5: Rename the column "Selling Price" to "Current Price" and "Original Price" to "MRP"

a.rename(columns={'Selling Price':"Current Price","Original Price":"MRP"},inplace=0)
```
